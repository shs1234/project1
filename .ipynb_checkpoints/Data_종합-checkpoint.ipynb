```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression #선형회귀 라이브러리의 선형회귀분석 모듈 
from sklearn import preprocessing  #데이터 전처리 사용
from sklearn.decomposition import PCA #PCA 분석
from sklearn.preprocessing import StandardScaler #표준화 작업
from  sklearn.preprocessing import PolynomialFeatures #다항식 변환
```

```python
import pandas as pd

#기온 데이터
df1=pd.read_excel('./data/기온_전체_2010-2019.xlsx')
#순서 역순으로 바꾸기
df1=df1.loc[::-1].reset_index(drop=True)
#평균기온, 평균최고, 평균최저만 남기고 drop
df1.drop(['지역번호','지역명','최고기온관측지점','최저기온관측지점','최고기온(℃)','최저기온(℃)'],axis=1,inplace=True)
#2019년 9월까지 자르기
temp=df1.iloc[96:108]
temp.set_index(['일시'],inplace=True)

#강수일수
df2=pd.read_excel('./data/2018_강수일수.xlsx')

df2T=df2.transpose()
rain_days=df2T[8].iloc[1:13]
rain_days.astype('float')

#강수량
rainfall=pd.read_excel('./data/2018_강수량.xlsx',skiprows=[1,2,3,4,5,6,7])
rainfall=rainfall['Unnamed: 2']

#황사일수
dust=pd.read_excel('./data/2018_황사일수.xlsx')
dustT=dust.transpose()
dust=dustT[8].iloc[1:13]

#체감온도
feelingtemp = pd.read_excel('./data/2018_체감온도.xlsx', header=3)

def month(day):
    return day[0:4] + '.' + day[5:7]

feelingtemp['월별'] = feelingtemp['일자'].astype(str).apply(month)
feelingtemp = feelingtemp.drop(['일자'], axis=1).groupby('월별').mean().drop(["기온(°C)",'풍속(m/s)'], axis=1)
feelingtemp

#소비자 물가지수
df7 = pd.read_csv('C:/project1/data/2018소비자물가지수.csv', encoding='cp949', header=0).iloc[:, 2:]
sobija = pd.DataFrame(df7.T[0])

#폭염지수
heat_wave = pd.read_excel('./data/2018_폭염지수.xlsx', header=5).iloc[[1], 1:13]
heat_wave = heat_wave.T

# 미세먼지
micro_dust=pd.read_excel("./data/2018_미세먼지농도.xlsx", header=5)
micro_dust['월별'] = micro_dust['일자'].astype(str).apply(month)
micro_dust = micro_dust.set_index('월별').drop(['지점번호', '지점명', '일자'], axis = 1)
micro_dust

# 구글 타임라인 월별 경복궁 검색
timeline=pd.read_excel('./data/multiTimeline (3).xlsx',skiprows=[0,1])
timeline.columns=['주', '검색 횟수']

timeline['Date']=pd.to_datetime(timeline['주'])
timeline['month']=timeline['Date'].dt.month

timeline=timeline['검색 횟수'].groupby(timeline['month']).sum()
timeline

#지하철승하차
subway=pd.read_excel('./data/서울교통공사 2018년 일별 역별 시간대별 승하차인원(1_8호선).xlsx',skiprows=1)
subway['month']=subway['날짜'].dt.month
subway_kbk=subway.loc[subway['역번호']==317, '합 계'].groupby(subway['month']).sum()


# 경복궁_관람객수
palace_visitors = pd.read_excel('./data/문화재청_궁능원_관람객_수_현황(2018년).xlsx', header=0).iloc[:,[0,3]]

palace_visitors['월별'] = palace_visitors['일자'].astype(str).apply(month)
palace_visitors = pd.DataFrame(palace_visitors.drop(['일자'], axis=1).groupby('월별').sum())

rain_days.index=rainfall.index=dust.index=feelingtemp.index=sobija.index=heat_wave.index=temp.index=subway_kbk.index=timeline.index=palace_visitors.index

df=pd.concat([temp,rain_days,rainfall,dust,feelingtemp,micro_dust, heat_wave,sobija,timeline, subway_kbk, palace_visitors],axis=1)
cols=['평균기온(℃)','평균최고기온(℃)','평균최저기온(℃)','강수일수','강수량','황사일수','체감기온(°C)','미세먼지농도(㎍/㎥)', '폭염지수','소비자물가지수', '검색 횟수', '경복궁역 승하차인원','경복궁_관람객수']
df.columns=cols
df
```

```python
#데이터 형태 확인 -꺽은선 그래프

f, axes = plt.subplots(2, 2, figsize=(15,10), sharex=True)

# 그래프 선명하게 보이게
import matplotlib
from IPython.display import set_matplotlib_formats
matplotlib.rc('font',family = 'Malgun Gothic')
set_matplotlib_formats('retina')

#기상그래프
xdf = df[['평균기온(℃)','평균최저기온(℃)','평균최고기온(℃)','체감기온(°C)']]
plt.rc('font',family='Malgun Gothic',size=7)
xdf.plot(marker=".",ax=axes[0, 0])

#기상일수+먼지 그래프
ydf = df[['강수일수','강수량','황사일수','폭염지수','미세먼지농도(㎍/㎥)']]
plt.rc('font',family='Malgun Gothic',size=7)
ydf.plot(marker=".",ax=axes[0, 1])

#소비자 물가지수
zdf = df[['소비자물가지수']]
plt.rc('font',family='Malgun Gothic',size=7)
zdf.plot(marker=".",ax=axes[1, 0])

#경복궁
z1df = df[['경복궁_관람객수']]
plt.rc('font',family='Malgun Gothic',size=7)
z1df.plot(marker=".",ax=axes[1, 1])
```

## 다중 다항 회귀분석

```python
# 훈련 데이터와 테스트 데이터로 분리
X = df[['소비자물가지수', '경복궁역 승하차인원', '검색 횟수']]   #독립변수 X
Y = df['경복궁_관람객수']    #종속변수 Y : 



# X 표준화
X = preprocessing.StandardScaler().fit(X).transform(X)

# train data 와 test data를 7:3 비율로 분리

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=10)  #랜덤 추출값 10

#다항식 변환
poly = PolynomialFeatures(degree=2)  # 2차항 적용
x_train_poly =poly.fit_transform(X_train)   # X학습데이터를 2차항으로 변형
print(x_train_poly.shape)
print(X_train.shape)

lr = LinearRegression()  #단순 선형회귀분석 객체 생성

lr.fit(x_train_poly, Y_train)  #학습


x_test_poly =poly.fit_transform(X_test) 
r_square = lr.score(x_test_poly, Y_test) # 결정 계수 계산
print(r_square )    #결정 계수 :    0.7087009262975685


print('기울기 a : ', lr.coef_)
print('절편 b : ', lr.intercept_)

#기울기 a :     [ 0.00000000e+00 -1.85768289e-02  1.70491223e-06]
# 절편 b :  62.58071221573144

y_test_predict = lr.predict(x_test_poly)  #test 데이터에 대한 모델의 예측값 구하기

# fig = plt.figure(figsize=(10, 5))
# ax = fig.add_subplot(1,1, 1)
# ax.plot(X_train, Y_train, '.', label='Train Data')
# ax.plot(X_test, y_test_predict, 'r+', label='Predicted Value')  #학습한 회귀선
# ax.legend(loc='best')
# plt.show()
# plt.close()

# X_poly = poly.fit_transform(X)
# y_predict = lr.predict(X_poly)   #전체 데이터 X에 대한 모델의 예측값 반환
# fig = plt.figure(figsize=(10, 5))
# ax1 = sns.distplot(Y, hist=False, label="Y")
# ax2 = sns.distplot(y_predict, hist=False, label="Y_predict", ax=ax1)
# plt.show()
# plt.close()

Y_prediction = y_test_predict.flatten()

for i in range(4):
    label = Y_test[i]
    prediction = Y_prediction[i]
    print("실제인원: %s, 예상인원: %s" %(label, prediction))
```
